```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
import gc
from statsmodels.tsa.stattools import adfuller
```

```

def pruebaDickeyFuller(datos):
  resultado = adfuller(datos)
  resultado = round(resultado[1], 3)
  return resultado

def obtenerDatasets():
  #Saltos en las columnas para obtener la métricas de calidad de enlace
  columnas = [i for i in range(14, 2710, 9)]
  for j in range(1, 11):
    #Recorrer y leer todos los archivos por cada fichero
    rutaArchivo = "/content/drive/MyDrive/PROYECTO -- EXPERIMENTOS/Script limpiar datasets/distance_10m/raw_data_run"+str(j) +".txt"
    data = pd.read_csv(rutaArchivo, header=None)
    cont = 1
    for i in columnas:
      data1 = data[[0, i]]

      #Selección de serie desde la columna de la métrica 
      serie = data1[[i]].values[:, -1]

      #Prueba Augmented Dickey-Fuller
      resp = pruebaDickeyFuller(serie)
      print("Analizando raw_data_run" + str(j) +"--" +str(i))

      # Valor de p>0.5 según la prueba de hipótesis para no estacionalidad
      if (resp >= 0.5):
        print("Ejecutado" + str(resp))

        #Reasignación de valores
        data1.loc[data1[i] == 0, i] = -1
        data1.loc[data1[i] > 0, i] = -1

        #Creación del archivo con la serie temporal en formato csv
        data1.columns = ["Date", "LQI"]
        data1.to_csv("/content/drive/My Drive/PROYECTO -- EXPERIMENTOS/Script limpiar datasets/Datset Limpios LQI/10m/raw_data_run" +
                     str(j) +"--" + str(cont) + ".csv", index = False)
        cont = cont + 1

    gc.collect()

    
```

```
obtenerDatasets()
```
